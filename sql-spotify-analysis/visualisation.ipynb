```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("results/average_features_per_genre")

df.plot(x='genre', y=['avg_danceability','avg_energy'], kind='bar')
plt.title("Average Audio Features by Genre")
plt.show()
```
